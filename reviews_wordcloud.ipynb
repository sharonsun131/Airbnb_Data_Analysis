```python
# Import our dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import warnings
% matplotlib inline
warnings.filterwarnings('ignore')
```

```python
# import Reviews csv for python word cloud
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
review_df = pd.read_csv('./Resources/reviews.csv' )
review_df.head(10)
```

```python
# Dropping columns
review_df.drop(columns=['listing_id', 'id', 'date', 'reviewer_id', 'reviewer_name'], inplace=True)
review_df.head()
```

```python
# Show obersavations and features
print("There are {} observations and {} features in this dataset. \n".format(review_df.shape[0],review_df.shape[1]))
```

```python
# Start with one review:
text = review_df.comments[1]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#review_df = ''.join(map(str, review_df))
review_df['comments'] = review_df['comments'].fillna(value='')
```

```python
# Bringing all text from the comments column
all_text = " ".join(comment for comment in review_df.comments)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["place", "Great", "recommended", "Thank", "stay", 'welcome', 'Highly', 'warm','wounderful', 'enjoyed', 'sister',
                 'every','min','able', 'want', 'made', 'lot', 'time', 'thing', 'br', 'much', 'need', 'needed', 'stayed', 'loved',
                 'day', 'plenty', 'everything', 'check', 'really', 'love', 'host', 'use', 'amenitie', 'recommend', 'will', 'unit',
                 'perfect', 'hosts', 'Airbnb', 'fantastic', 'enjoy', 'one', 'feel', 'trip', 'find', 'u', 'definitely', 'us', 'night',
                 's', 'make', 'come back', 'chair', 'staying'])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="black", max_words=35, max_font_size=60).generate(all_text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from os import path
from PIL import Image
```

```python
# Creating a Mask for our WordCloud
mask =  np.array(Image.open('img/hawaii_islands.jpg'))
mask
```

```python
# Generating a WordCloud
wc = WordCloud(background_color="white", max_words=200, mask=mask,
    stopwords=stopwords, colormap='tab10')

# generate wordcloud
wc.generate(all_text)

plt.figure(figsize=[50,25])
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Generating a WordCloud
wc_dark = WordCloud(background_color="black", max_words=500, mask=mask,
    stopwords=stopwords, colormap='tab10')

# generate wordcloud
wc_dark.generate(all_text)

plt.figure(figsize=[50,25])
plt.imshow(wc_dark, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Generating a WordCloud
wc_new = WordCloud(background_color="white", max_words=200, mask=mask,
    stopwords=stopwords, colormap='magma')

# generate wordcloud
wc_new.generate(all_text)

plt.figure(figsize=[50,25])
plt.imshow(wc_new, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Generating a WordCloud
wc_color = WordCloud(mode = 'RGB', background_color='rgb(12, 186, 166)', max_words=1000, mask=mask,
    stopwords=stopwords, colormap='tab10')

# generate wordcloud
wc_color.generate(all_text)

plt.figure(figsize=[50,25])
plt.imshow(wc_color, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Generating a WordCloud
wc_color_new = WordCloud(mode = 'RGB', background_color='rgb(0, 84, 119)', max_words=500, mask=mask,
    stopwords=stopwords, colormap='tab10', contour_width=0.5, contour_color='whitesmoke')

# generate wordcloud
wc_color_new.generate(all_text)

plt.figure(figsize=[50,25])
plt.imshow(wc_color_new, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# new wc image with background
image_colors = ImageColorGenerator(mask)

wc_background = WordCloud(stopwords=stopwords, 
                        background_color="white", 
                        mode="RGBA", 
                        max_words=1000, 
                        #contour_width=3, 
                        repeat=True,
                        mask=mask,
                        colormap='summer')

wc_background.generate(all_text)
wc_background.recolor(color_func = image_colors)

plt.figure(figsize=[50,25])
plt.imshow(wc_background)
plt.axis("off")
plt.show()
```

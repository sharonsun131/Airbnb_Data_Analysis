```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
neighborhoods_df=gpd.read_file('neighbourhoods.geojson')
neighborhoods_df.head()
```

```python
neighborhoods_df['geometry'][0]
```

```python
listing_neighborhoods_df=pd.read_csv('listings.csv')
```

```python
listing_neighborhoods_df=listing_neighborhoods_df[['id','neighbourhood_cleansed','neighbourhood_group_cleansed','latitude','longitude']]
listing_neighborhoods_df.head()
```

```python
listing_neighborhoods_df.isnull().values.any()
```

```python
fig = px.scatter_geo(listing_neighborhoods_df,lat='latitude',lon='longitude', hover_name="id")
fig.update_layout(title = 'Listings by Neighborhood', title_x=1)
fig.show()
```

```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt

```

```python

import warnings
warnings.filterwarnings("ignore")
```

```python
reviews_df=pd.read_csv('reviews-3.csv')
reviews_df.head()
```

```python
reviews_df=reviews_df.drop('date', axis=1)
reviews_df.head()
```

```python
reviews_df=reviews_df.drop('id', axis=1)
reviews_df.head()
```

```python
reviews_df=reviews_df.drop('reviewer_id', axis=1)
reviews_df.head()
```

```python
reviews_df.isnull().values.any()
```

```python
reviews_df=reviews_df.dropna()
reviews_df.isnull().values.any()
```

```python
reviews_df["comments"] = reviews_df["comments"].fillna(value="")
```

```python
text = " ".join(reviews_df['comments'].str.lower())
stopwords = ['br','property','room','place','condo','apartment','check','house','u'] + list(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="black", max_words=30).generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
reviews_df.to_csv('reviews_comments.csv', sep='\t', encoding='utf-8', index=False)
```
